```r
# 추가적으로 layout에 대한 설명
op = par(no.readonly = TRUE)
# par함수에 비해 사용자에게 보다 높은 자유도를 제공하므로, 매우 유용하고 강력한 함수이다.
layout(matrix(c(1,2,3,4), 2, 2, byrow= TRUE))
layout.show(4)
# layout.show(n)은 현재의 layout에 대한 외곽선을 n 만큼의 그래프 갯수만큼 볼 수 있게 해주는 함수이다.
```

```r
layout(matrix(c(1,0,2,2), 2, 2, byrow =TRUE))
layout.show(2)
```

```r
layout(matrix(c(1,2,3,3), 2, 2, byrow=TRUE))
layout.show(3)
```

```r
l = layout(matrix(1), widths = lcm(5), heights = lcm(10))
# 너비가 5이고 높이가 10인 직사각형
layout.show(1)
```

```r
layout(matrix(c(2,0,1,3), 2, 2, byrow= TRUE),
      widths = lcm(c(8,4)),
      heights = lcm(c(4,8)),
      respect = TRUE)
# respect = TRUE는 가로 넓이와 세로 높이의 비율을 고려해서 그래프 영역을 설정
layout.show(3)
```
